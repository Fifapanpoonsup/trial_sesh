## Splitting data

```python
def data_split(BRK_close_price,window_size):
    data_close_conglomerate = BRK_close_price
    data_conglomerate =[]
  
    for index in range(len(data_close_conglomerate)-window_size):
        data_conglomerate.append(data_close_conglomerate[index:index+window_size])
    data_con = np.array(data_conglomerate)
    test_set_size = int(np.round(0.2*data_con.shape[0]))
    train_set_size = data_con.shape[0] - (test_set_size)
    
    x_train_BRK = data_con[:train_set_size,:-1,:]
    y_train_BRK = data_con[:train_set_size,-1,:]
    
    x_test_BRK = data_con[train_set_size:,:-1]
    y_test_BRK = data_con[train_set_size:,-1,:]
    
    return [x_train_BRK, y_train_BRK, x_test_BRK, y_test_BRK]
window_size = 7
x_train_BRK,y_train_BRK,x_test_BRK,y_test_BRK = data_split(BRK_close_price,window_size)
print('x_train.shape = ',x_train_BRK.shape)
print('y_train.shape = ',y_train_BRK.shape)
print('x_test.shape = ',x_test_BRK.shape)
print('y_test.shape = ',y_test_BRK.shape)
```

```python

class LSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, output_dim):
        super(LSTM, self).__init__()
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim).requires_grad_()
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim).requires_grad_()
        out, (hn, cn) = self.lstm(x, (h0.detach(), c0.detach()))
        out = self.fc(out[:, -1, :]) 
        return out
```

```python
model = LSTM(input_dim=input_dim, hidden_dim=hidden_dim, output_dim=output_dim, num_layers=num_layers)
criterion = torch.nn.MSELoss(reduction='mean')
optimiser = torch.optim.Adam(model.parameters(), lr=0.01)
```

## LSTM MODEL

```python
class lstm(nn.Module):
    def __init__(self, output_dim,input_size,hidden_size, num_layers,seq_len):
        super(lstm,self).__init__()
        self.output_dim = output_dim# number of class
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers 
        self.seq_len = seq_len
        
        self.lstm = nn.LSTM(input_size = input_size,hidden_size = hidden_size, num_layers = num_layers,batch_first = True)
        self.fc1 = nn.Linear(hidden_size , 128)
        self.fc = nn.Linear(128, output_dim )
        self.relu = nn.ReLu()
    def forward(self,x):
        h0 = Variable(torch.zeros(self.num_layers,x.size(0),self.hidden_size))# hidden
        c0 = Variable(torch.zeros(self.num_layers,x.size(0),self.hidden_size))#internal 
        output, (hn,cn) = self.lstm(x,(h0,c0))
        hn = hn.view(-1,self.hidden_size)# reshaping the data from one dense layer to another 
        out = self.relu(hn)
        out = self.fc1(out)
        out = self.relu(out)
        out = self.fc(out)
        return out
        
```
